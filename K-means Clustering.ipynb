```python
sn.pairplot(data=df)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2,21):
    knn=KMeans( n_clusters=i, init='k-means++',)
    knn.fit(df)
    wcss.append(knn.inertia_)
```

```python
# help(KMeans)
```

```python
plt.plot([i for i in range(2,21)],wcss,marker='o')
plt.xticks([i for i in range(2,21)])
plt.grid(axis='x')
plt.show()
```

```python
plt.plot([i for i in range(2, 21)], wcss, marker='o')
plt.xticks([i for i in range(2, 21)])
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal k')
plt.grid(axis='x')
plt.show()
```

```python
wcss = []
for i in range(1, 21):
    knn = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    knn.fit(df)
    wcss.append(knn.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 5))
plt.plot(range(1, 21), wcss, marker='o')
plt.xticks(range(1, 21))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
knn=KMeans(n_clusters=3)
df['predict']=knn.fit_predict(df)
df
```

```python
sn.pairplot(data=df,hue='predict')
```
